```python
cnt = 0

inn = None
kompleks = None
adress = None
class_nedvizh = None
otdelka = None
id_house = None
data_sdachi = None
data_sdachi_vspom = None

id_house_soup = []
id_houses = []
class_nedvizh_list = []
data_sdachi_list = []
data_sdachi_list_vspom = []

df = pd.DataFrame(columns=['ИНН', 'ID_House', 'Наим_комплекса', 'Адресс', 'Класс_недвиж', 'Отделка', 'Дата_сдачи']) 
```

```python
with open(r'C:\Users\Михайлова Оксана\Desktop\Paser_nash_dom_rf\inn.txt', 'r') as f:
    list_inn = f.read()
    list_inn_split = list_inn.split('\n')

driver = webdriver.Chrome(r'C:\webdrivers\chromedriver.exe')
url_inn = 'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/%D0%BA%D0%B0%D1%82%D0%B0%D0%BB%D0%BE%D0%B3-%D0%BD%D0%BE%D0%B2%D0%BE%D1%81%D1%82%D1%80%D0%BE%D0%B5%D0%BA/%D1%81%D0%BF%D0%B8%D1%81%D0%BE%D0%BA-%D0%BE%D0%B1%D1%8A%D0%B5%D0%BA%D1%82%D0%BE%D0%B2/'
url_stroi_object = 'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/%D0%BA%D0%B0%D1%82%D0%B0%D0%BB%D0%BE%D0%B3-%D0%BD%D0%BE%D0%B2%D0%BE%D1%81%D1%82%D1%80%D0%BE%D0%B5%D0%BA/%D0%BE%D0%B1%D1%8A%D0%B5%D0%BA%D1%82/'

```

```python
for i in list_inn_split:
    try:
        inn = i
        cnt += 1
        print(cnt, inn)

        driver.get(url_inn)

        element = driver.find_element(By.XPATH, '/html/body/div[2]/div/div[1]/div/div/div[2]/div[2]/div[1]/div[1]/div/div[1]/div[1]/div/input')
        element.clear()
        element.send_keys(inn)
        element.send_keys(Keys.ENTER)
        time.sleep(0.5)
        pageSource = driver.page_source
        soup = BeautifulSoup(pageSource)
        id_house_soup = soup.find_all('span',class_="styles__Ellipsis-sc-1fw79ul-0 cDcbYl styles__Value-sc-1ajig0k-5 styles__IdValue-sc-1ajig0k-17 grqFVa")

        for i in id_house_soup:
            a = i.text     
            a = a[a.find('ID дома:') + 9 : len(a)]
            id_houses.append(a)

        id_houses = list(set(id_houses))
        
        data_sdachi_list = soup.find_all('div',class_="styles__Cell-sc-7809tj-0 ibavEN Newbuindings BuildersTable_normal")

        for i2 in range(len(data_sdachi_list)):
            b = data_sdachi_list[i2].text
            if 'кв.' in b:
                data_sdachi_vspom = b
#                print(data_sdachi_vspom)
                data_sdachi_list_vspom.append(data_sdachi_vspom)        
                        
#         print(id_houses)   
#         print(len(id_houses))
#         print(len(data_sdachi_list_vspom))         

        for i1 in range(len(id_houses)):
            id_house = id_houses[i1]
            data_sdachi = data_sdachi_list_vspom[i1]
#            print(data_sdachi)          
            driver.get(url_stroi_object + id_house)
            time.sleep(0.5)
            pageSource = driver.page_source
            soup1 = BeautifulSoup(pageSource)

            kompleks = soup1.find('h1',class_="styles__Name-sc-eng632-3 kGUCcv").text
            adress = soup1.find('p',class_="styles__Address-sc-eng632-11 ljfwme").text
            
            class_nedvizh_list = soup1.find_all('span',class_="styles__RowSpan-sc-1fyyfia-4 dAtYoX")
            class_nedvizh = class_nedvizh_list[1].text
            otdelka = class_nedvizh_list[5].text
                    
            df = df.append({'ИНН':inn, 'ID_House':id_house, 'Наим_комплекса':kompleks, 'Адресс':adress,
                            'Класс_недвиж':class_nedvizh, 'Отделка':otdelka, 'Дата_сдачи':data_sdachi}
                            , ignore_index=True)    

        inn = None
        kompleks = None
        adress = None
        class_nedvizh = None
        otdelka = None
        data_sdachi = None
        data_sdachi_list = None

        id_house_soup = []
        id_houses = []
        class_nedvizh_list = []
        data_sdachi_list = []
        data_sdachi_list_vspom = []
        
        
    except: 
        print('косяк', cnt, inn)
        inn = None
        kompleks = None
        adress = None
        class_nedvizh = None
        otdelka = None
        data_sdachi = None
        data_sdachi_list = None
        
        id_house_soup = []
        id_houses = []
        class_nedvizh_list = []
        data_sdachi_list = []
        data_sdachi_list_vspom = []

    
```
